```python
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
# Distributions: 
vis1 = sns.distplot(stats["InternetUsers"])    # distplot-Distribution plot, it is univariate analysis because plot graph using 1 var
```

```python
vis1 = sns.displot(stats["InternetUsers"])
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=10)    # 10 bins created
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=20)    # 20 bins(graph columns) created
```

```python
#BOX PLOTS:
vis2 = sns.boxplot(data = stats, x="IncomeGroup", y='BirthRate')    # Bi-variate analysis i.e. uses 2 variables
```

```python
# refer to seaborn gallary 
```

```python
# visualizing with seaborn
```

```python
vis4 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate')    # Bydefault fit_reg = True
```

```python
vis3 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model, (bi-variate)
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = True,hue = 'IncomeGroup')
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup', size = 10)
```
